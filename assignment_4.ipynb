For this Seaborn exercise, I chose to use the dataset [Kaggle: Wine Reviews](https://www.kaggle.com/datasets/zynicide/wine-reviews?datasetId=1442&searchQuery=seaborn&select=winemag-data_first150k.csv).

```python
# import libraries
import pandas as pd
import seaborn as sns
sns.set(style="whitegrid")
```

```python
# import data
df = pd.read_csv('winemag-data_first150k.csv')
```

## Data Visualization using Seaborn

```python
# check the distribution of review points using countplot
sns.countplot(df['points'])
```

```python
# check the distribution of price of a bottle of the wine using kdeplot
sns.kdeplot(df['price'])
```

```python
# seems like power law distribution? need to zoom in to see the lower price range
sns.kdeplot(df[df['price']<200]['price'])
```

```python
# Check the distributions for two dimensions incl. points and price using joint hexplot
sns.jointplot(x='price', y='points', data=df[df['price'] < 100], kind='hex', gridsize=20)
# Most observations are for wine with price between 15 and 20, and review points between 86 and 88
```

```python
# Check review points distributions for the top most reviewed wine varieties using boxplot
ax = sns.boxplot(x='variety',y='points',data=df[df['variety'].isin(df['variety'].value_counts().head(10).index)])
ax.set_xticklabels(ax.get_xticklabels(),rotation=90)
# Bordeaux-style Red Blend has the highest median review points
```
